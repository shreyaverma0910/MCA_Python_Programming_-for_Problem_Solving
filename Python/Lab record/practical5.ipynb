```python
def reverse_vowels(s):
    vowels = "aeiouAEIOU"
    s = list(s)

    # Collect vowels
    v = [ch for ch in s if ch in vowels]

    # Reverse the vowels list
    v.reverse()

    # Put reversed vowels back into string positions
    idx = 0
    for i in range(len(s)):
        if s[i] in vowels:
            s[i] = v[idx]
            idx += 1

    return "".join(s)

# Input
text = input("Enter sentence: ")
print("Enter sentence:",text)

# Output
print(reverse_vowels(text))
```
