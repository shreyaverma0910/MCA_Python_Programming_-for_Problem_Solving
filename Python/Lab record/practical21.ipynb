```python
import pandas as pd

# Create the sales dataset
data = {
    "Order_ID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010,1011,1012,1013,1014,1015],
    "Product": ["Notebook","Laptop","Pen Set","Mouse","Keyboard","Printer",
                "Marker Set","Headphones","USB Drive","Scanner","Calculator",
                "Webcam","Tablet","File Folder","Monitor"],
    "Category": ["Stationery","Electronics","Stationery","Accessories","Accessories","Electronics",
                 "Stationery","Accessories","Accessories","Electronics","Stationery",
                 "Accessories","Electronics","Stationery","Electronics"],
    "Quantity": [50,2,100,15,12,3,80,20,35,5,40,8,4,120,6],
    "Price": [120,45000,250,350,800,8500,180,1200,400,6500,350,2500,25000,150,12000],
    "Month": ["January","January","January","February","February","February",
              "February","March","March","March","April","April","April","April","April"]
}

# Load into DataFrame
df = pd.DataFrame(data)

# Create Total_Revenue column
df["Total_Revenue"] = df["Quantity"] * df["Price"]

# Total revenue for each product
print("Total revenue for each product:")
print(df.groupby("Product")["Total_Revenue"].sum())

# Total units sold for each category
print("\nTotal units sold for each category:")
print(df.groupby("Category")["Quantity"].sum())

# Average price per unit for each category
print("\nAverage price per unit for each category:")
print(df.groupby("Category")["Price"].mean())

# Product with maximum units sold
print("\nProduct with maximum units sold:")
print(df.loc[df["Quantity"].idxmax()])

# Total revenue per month
print("\nTotal revenue for each month:")
print(df.groupby("Month")["Total_Revenue"].sum())

# Products with Units_Sold greater than 40
print("\nProducts with Units Sold greater than 40:")
print(df[df["Quantity"] > 40])

# Sort dataset by Total_Revenue in descending order
print("\nDataset sorted by Total_Revenue (Descending):")
print(df.sort_values(by="Total_Revenue", ascending=False))
```
