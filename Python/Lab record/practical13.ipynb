```python
import random
import string

# Character sets
uppercase = string.ascii_uppercase
lowercase = string.ascii_lowercase
digits = string.digits
special_symbols = "!@#$*"

# First character must be uppercase or digit
first_char = random.choice(uppercase + digits)

# Remaining characters
all_chars = uppercase + lowercase + digits + special_symbols
remaining_chars = random.choices(all_chars, k=11)

# Combine to form password
password = first_char + "".join(remaining_chars)

print("Generated Password:", password)
```
