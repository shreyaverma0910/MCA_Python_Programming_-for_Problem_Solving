```python
import math

# Taking input from the user
num = float(input("Enter a number: "))

#  Square root of the number
square_root = math.sqrt(num)
print("Square root of", num, "is:", square_root)

#  Factorial of the number
if num >= 0 and num.is_integer():
    factorial_value = math.factorial(int(num))
    print("Factorial of", int(num), "is:", factorial_value)

#  Power of a number
power = float(input("Enter the power value: "))
result = math.pow(num, power)
print(num, "raised to the power", power, "is:", result)
```
