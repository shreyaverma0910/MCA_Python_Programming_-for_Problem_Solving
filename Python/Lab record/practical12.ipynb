```python
import random

# List of students
students = [
     "Priya", "Karan", "Shalini", "Rocky", "Anjali"
]

# Randomly select a CR
cr = random.choice(students)
print("Selected CR:", cr)

# Shuffle the student list
random.shuffle(students)

# Print the shuffled list
print("Shuffled student list:")
for student in students:
    print(student)
```
