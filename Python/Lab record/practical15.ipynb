```python

numbers = [10, 20, 30, 40, 50]

try:
    index = int(input("Enter an index: "))
    print("Element at index", index, "is:", numbers[index])

except IndexError as e:
    print("IndexError:", e)

except TypeError as e:
    print("TypeError:", e)

except ValueError as e:
    print("ValueError:", e)
```
