```python
#pandas alias 
import pandas as pd
data={
    "Name":["Amit","Aman","Sohan"],
    "Age":[24,36,67],
    "Marks":[87,99,67]
}
df=pd.DataFrame(data)
print("Dataframe:")
print(df)
#selecting a single column
print("\nNames column:")
print(df["Name"])
#calculating averge marks
print("\nAverage Marks:")
print(df["Marks"].mean())
```

```python
#importing all functions
from math import *

num = 16

print("Square root:", sqrt(num))
print("Factorial:", factorial(3))
print("Value of pi:", pi)
print("Cosine of 0:", cos(0))
print("Power (2^4):", pow(2, 3))
```

```python
#rando modules
import random

# 1. random.random()
print("Random float (0 to 1):", random.random())

# 2. random.randint(a, b)
print("Random integer (1 to 10):", random.randint(1, 10))

# 3. random.randrange(start, stop)
print("Random number from range (5 to 20):", random.randrange(5, 20))

# 4. random.choice(list)
fruits = ['apple', 'banana', 'mango', 'orange']
print("Random choice from list:", random.choice(fruits))

# 5. random.shuffle(list)
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print("List after shuffle:", numbers)

# 6. random.uniform(a, b)
print("Random float between 10 and 20:", random.uniform(10, 20))
```
