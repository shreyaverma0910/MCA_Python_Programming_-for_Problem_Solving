```python
#Array of Zeros and Ones
import numpy as np
print(np.zeros((2,3)))
print(np.ones((3,4)))
```

```python
import numpy as np
#Creating NumPy Arrays
print(np.array([1,2,3]))     
#2D Array
matrix=np.array([[1,2],[3,4]])
print("2D matrix:")
print(matrix)
#matrix operations
a=np.array([[2,3],[4,5]])
b=np.array([[5,6],[7,8]])
print("Matrix addition:\n",a+b)
print("Matrix multiplication:\n",a*b)
print("Transpose:\n",a.T)
#Diagonal Elements (2D Array)
diag=np.diag(matrix)
print("Diagonal:",diag)
```

```python
#Array Range and Linespace
import numpy as np
print(np.arange(2,20,3)) 
print(np.arange(5,20))
print(np.linspace(3,30,10))
#reshape operations
a=np.arange(1,10)
print(a.reshape(3,3))
```

```python
import numpy as np 
#Arithmetic operations
a=np.array([10,20,30])
b=np.array([1,2,3])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
#mathemetical functions
print(np.sqrt(25))
print(np.pow(2,5))
```

```python
import numpy as np
#Statistical Operations
arr=np.array([45,20,50,70,65])
print("Mean",np.mean(arr))
print("Median:",np.median(arr))
print("Sum:",np.sum(arr))
print("Std dev:",np.std(arr))
```

```python
#Indexing and Slicing 
import numpy as np 
arr=([10,20,30,40,50,60,70,80,90])
print("Index element:",arr[5])
print(arr[:6])
print(arr[7:])
print(arr[-4])
print(arr[::3])
print(arr[5::])
print(arr[1:7])
```

```python
#multiples
import numpy as np 
arr=np.array([2,3,4,5,6,7,8,9,10,12,14,15])
multiples=arr[arr%3==0]
print("Multiple of 3:",multiples)
multiples=arr[arr%5==0]
print("Multiple of 5:",multiples)
multiples=arr[arr%4!=0]
print("Not a multiple of 4:",multiples)
even=arr[arr%2==0]
print("Even numbers are:",even)
odd=arr[arr%2!=0]
print("Odd numbers are:",odd)

#greater and smaller
greater = arr[arr > 8]
print("Greater numbers are:",greater)
smaller = arr[arr < 8]
print("Smaller numbers are:",smaller)
```
