```python
#Q5. pass,break and continue statements
for i in range(1,6):
    if i==2:
        pass
    elif i==3:
        continue
    elif i==5:
        break
    print("Current Value:",i)
print("Loop ended.")
```

```python
#Q6. calculate compound interest
def compound_interest(p,r=5,t=1):
    return p*(1+r/100)**t-p
ci=compound_interest(1000,5,2)
print("Compound interest:",ci)
```

```python
#Q7. nested control structure
for i in range(1,6):
    for j in range(1,11):
        if i==3 or j==3:
            continue
        print(i,"x",j,"=",i*j)
```

```python
#Q8. set and dictionary
#set demonstration 
A={1,2,3,4}
B={3,4,5,6}
print("Union:",A|B)
print("Intersection:",A&B)
print("Diiference:",A-B)

#dictionary demontration
student={"name":"shreya","age":20,"course":"MCA"}
student["city"]="Delhi"
student["age"]=21
del student["course"]
print("Dictionary:",student)
```

```python
#Q9. string slicing and manipulation techniques
name=input("Enter full name:")
if name=="":
    print("Error:Empty input is not allowed")
else:
    words=name.split()
    
    initials=""
    for w in words:
        initials=initials+w[0].upper()

    reverse_name=name[::-1]
    title_name=name.title()
    print("Initials:",initials)
    print("Reversed name:",reverse_name)
    print("Title case:",title_name)
```

```python
#Q10. file handling 
try:
    file=open("numbers.txt","r")
    numbers=file.read().split()
    file.close()

    total=0
    for n in numbers:
        total=total+int(n)
    average=total/len(numbers)

    file=open("result.txt","w")
    file.write("Sum:"+{total}+"\n")
    file.write("Average:"+{average}+"\n")
    file.close()

    print("Result return to result.txt")
    
except FileNotFoundError:
    print("File not found")
               


    
```
