```python
#Guessing number

secret_number=7
attempts=5

for i in range(attempts):
    guess=int(input("Enter guessing number: "))

    # if not guess.isdigit():
    #     print("Please enter a valid number")
    #     continue

    # guess = int(guess)

    if guess==secret_number:
        print("Correct guess")
        break
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Too low!")

else:
    print("Sorry too many attempts and correct number is 7")
  


    
```
