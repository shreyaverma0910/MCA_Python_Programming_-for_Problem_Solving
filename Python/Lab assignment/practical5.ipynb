```python
employees = [
    {"name": "John", "salary": 45000},
    {"name": "Mary", "salary": 52000},
    {"name": "Alex", "salary": 48000}
]

print("Original Employee list:")
for emp in employees:
    print(emp)

# Sort using salary as the key
sorted_employees = sorted(employees, key=lambda x: x["salary"])

print("\nSorted by Salary:")
for emp in employees:
    print(emp)
```
