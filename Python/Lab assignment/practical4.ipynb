```python
# Accept list from user (comma-separated values)
user_input = input("Enter list items separated by commas: ")
print("Enter elements:",user_input)
lst = list(map(int, user_input.split(",")))

# Accept item to search
item = int(input("Enter item to search: "))

# Search for the item
found = False
for i in range(len(lst)):
    if lst[i] == item:
        print(f"Item found at position {i+1}")
        found = True
        break

if not found:
    print("Item not found.")
```
