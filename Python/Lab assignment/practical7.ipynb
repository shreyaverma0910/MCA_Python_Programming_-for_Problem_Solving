```python
import string

text = "Rain rain go away, come again another day."
print("Input text:",text)

# Convert to lowercase
text = text.lower()

# Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# Split into words
words = text.split()

# Count frequency
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1
print("\nWord Frequency:")
print(freq)
```
