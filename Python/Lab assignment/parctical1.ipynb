```python
# Pattern 1:Number triangle 
print("Pattern 1:\n")
for i in range(1,6):
    for j in range(1,i+1):
        print(j,end="")
    print()
    
print()

#Pattern 2: Diamond-star like pattern
rows = 4
print("Pattern 2:\n")
for i in range(rows):
    print(" " * (rows - i - 1), end=" ")    
    print("*" * (2*i + 1))

#Pattern 3: Bigger diamond Pattern
rows = [1, 3, 5, 3, 1]
print("Pattern 3:\n")
for r in rows:
    spaces = (5 - r) // 2
    print("  " * spaces, end="")
    for i in range(r):
        print("*", end=" ")
    print()

#Pattern 4: Square border pattern 
rows = 4
cols = 3
print("Pattern 4:\n")
for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()

#Pattern 5: Cross star pattern
n = 3
print("Pattern 5:\n")
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
